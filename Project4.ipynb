```python
import numpy
import pandas
import matplotlib.pyplot as PLT

Data = [Line.replace('\n', '').split(',') for Line in open('reservior1.csv', 'rt').readlines()]
```

```python
head = Data.pop(0)
print(head)
```

```python
plotdata = list()
for line in Data:
    date, poondi, cholavaram, redhills, chembaramkkam = line
    record = [date, int(poondi), int(cholavaram), int(redhills), int(chembaramkkam)]
    plotdata.append(record)

MyData = pandas.DataFrame(plotdata, columns = head)
print(MyData)
```

```python
import matplotlib.pyplot as PLT
PLT.subplot(221)
PLT.plot(MyData['Date'][1:], MyData['POONDI'][1:], label = "Poondi")
PLT.subplot(222)
PLT.plot(MyData['Date'][1:], MyData['CHOLAVARAM'][1:], label = "Cholavaram")
PLT.subplot(223)
PLT.plot(MyData['Date'][1:], MyData['REDHILLS'][1:], label = "Redhills")
PLT.subplot(224)
PLT.plot(MyData['Date'][1:], MyData['CHEMBARAMBAKKAM'][1:], label = "Chembarambakkam")
PLT.show()
```
