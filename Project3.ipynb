```python
import numpy
import pandas
import matplotlib.pyplot as PLT

Data = [Line.replace('\n', '').split(',') for Line in open('AirImport.csv', 'rt').readlines()]
```

```python
header = Data.pop(0)
head = header[1:]
print(head)
```

```python
plotdata = list()
for line in Data[0:27]:
    code, product, year1213, year1314 = line
    record = [product, float(year1213), float(year1314)]
    plotdata.append(record)

MyData = pandas.DataFrame(plotdata, columns = head)
print(MyData)
```

```python
import matplotlib.pyplot as PLT
PLT.subplot(121)
PLT.plot(MyData['Principal Commodity through MUMBAI AIR'][1:], MyData['2012-13 (Value in Rs. Crore)'][1:])
PLT.subplot(122)
PLT.plot(MyData['Principal Commodity through MUMBAI AIR'][1:], MyData['2013-14 (Value in Rs. Crore)'][1:])
```
