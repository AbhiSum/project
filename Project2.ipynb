```python
Data = [Line.replace('\n', '') for Line in open('matches.csv', 'rt').readlines()]
```

```python
Data.pop(0)
Data
```

```python
twffw = 0
twff = 0
twbfw = 0
twbf = 0
tw = 0
tlffw = 0
tlff = 0
tlbfw = 0
tlbf = 0
tl = 0
match = 0
for line in Data:
    record = line.split(',')
    if record[2] == 'Royal Challengers Bangalore' or record[3] == 'Royal Challengers Bangalore':
        if record[4] == 'Royal Challengers Bangalore':
            if record[5] == 'field': 
                if record[8] == 'Royal Challengers Bangalore':
                    twffw = twffw + 1
                twff = twff + 1
            else:
                if record[8] == 'Royal Challengers Bangalore':
                    twbfw = twbfw + 1
                twbf = twbf + 1
            tw = tw + 1
        else:
            if record[5] == 'field': 
                if record[8] == 'Royal Challengers Bangalore':
                    tlffw = tlffw + 1
                tlff = tlff + 1
            else:
                if record[8] == 'Royal Challengers Bangalore':
                    tlbfw = tlbfw + 1
                tlbf = tlbf + 1
            tl = tl + 1

        match = match + 1
        
    
print('Chance of RCB winning match if toss won and field first', twffw / twff)
print('Chance of RCB winning match if toss won and bat first', twbfw / twbf)
print('Chance of RCB winning match if toss lost and field first', tlffw / tlff)
print('Chance of RCB winning match if toss lost and bat first', tlbfw / tlbf)
print('Chance of RCB winning match if toss won', tw / match)
print('Chance of RCB winning match if toss lost', tl / match)
print('Chance of RCB winning match if field first', (twffw + tlffw) / (twff + tlff))
print('Chance of RCB winning match if bat first', (twbfw + tlbfw) / (twbf + tlbf))
```
