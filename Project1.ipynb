```python
Data = [Line.replace('\n', '') for Line in open('rainfall.csv', 'rt').readlines()]
```

```python
Data
Data.pop(0)
Data
```

```python
i = 2
ann = 0
AData = list()
for line in Data:
    Record = line.split(',')
    while i < 14:
        ann = ann + float(Record[i])
        i = i + 1
    i = 2
    AData.append((Record[1], ann))
    ann = 0
AData
```

```python
for x in AData:
    avg = x[1] / 12
    print(avg)
```
